```python
import os
import pandas as pd

TSV_FOLDER = "/content/drive/MyDrive/CS406/Project/Dataset/boxes_and_transcripts"
OUTPUT_FOLDER = "/content/drive/MyDrive/CS406/Project/Dataset/processed_tsv"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

TARGET_COLS = 11  # số cột chuẩn

def process_file(file_path):
    rows = []

    with open(file_path, "r", encoding="utf-8") as f:
        for line_no, line in enumerate(f, start=1):
            cols = line.rstrip("\n").split(",")

            # gộp từ phải sang trái, trừ cột label
            while len(cols) > TARGET_COLS:
                left = -3   # trước label
                right = -2  # trước label 1 cột

                if cols[left] and cols[right]:
                    cols[left] = cols[left].rstrip() + " " + cols[right].lstrip()
                elif not cols[left]:
                    cols[left] = cols[right]

                cols.pop(right)

            rows.append(cols)

    return pd.DataFrame(rows)

# xử lý toàn bộ thư mục
for fname in os.listdir(TSV_FOLDER):
    if fname.endswith(".tsv") or fname.endswith(".csv"):
        in_path = os.path.join(TSV_FOLDER, fname)
        out_path = os.path.join(OUTPUT_FOLDER, fname)

        df = process_file(in_path)

        # assert an toàn
        bad = df[df.apply(lambda r: len(r) != TARGET_COLS, axis=1)]
        if not bad.empty:
            raise ValueError(f"{fname} vẫn còn dòng != {TARGET_COLS} cột")

        df.to_csv(out_path, index=False, header=False)
        print(f"✅ Processed: {fname}")
```

```python
import os

INPUT_FOLDER = "/content/drive/MyDrive/CS406/Project/Dataset/processed_tsv"
OUTPUT_FOLDER = "/content/drive/MyDrive/CS406/Project/Dataset/processed_bio"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

TARGET_COLS = 11  # cột cuối là label

def is_new_entity(prev_text, curr_text):
    # dòng đầu hoặc dòng trước rỗng
    if not prev_text:
        return True

    # anchor invoice thường kết thúc bằng :
    if curr_text.endswith(":"):
        return True

    # text khác nhau → entity mới
    if curr_text != prev_text:
        return True

    return False


for fname in os.listdir(INPUT_FOLDER):
    if not (fname.endswith(".csv") or fname.endswith(".tsv")):
        continue

    in_path = os.path.join(INPUT_FOLDER, fname)
    out_path = os.path.join(OUTPUT_FOLDER, fname)

    prev_label = "OTHER"
    prev_text = ""

    with open(in_path, "r", encoding="utf-8") as fin, \
         open(out_path, "w", encoding="utf-8") as fout:

        for line in fin:
            cols = line.rstrip("\n").split(",")

            if len(cols) != TARGET_COLS:
                continue

            text = cols[-2].strip()
            label = cols[-1].strip()

            if label == "OTHER":
                bio_label = "O"
            else:
                if label != prev_label or is_new_entity(prev_text, text):
                    bio_label = f"B-{label}"
                else:
                    bio_label = f"I-{label}"

            cols[-1] = bio_label
            fout.write(",".join(cols) + "\n")

            prev_label = label
            prev_text = text

    print(f"✅ BIO processed: {fname}")
```

```python
import os
import csv
from PIL import Image
from sklearn.model_selection import train_test_split

IMAGE_FOLDER = "/content/drive/MyDrive/CS406/Project/Dataset/train_images"
DATA_FOLDER = "/content/drive/MyDrive/CS406/Project/Dataset/processed_tsv"

LABELS = [
    "OTHER",
    "ADDRESS",
    "TOTAL_COST",
    "TIMESTAMP",
    "SELLER",
]

label2id = {l: i for i, l in enumerate(LABELS)}
id2label = {i: l for l, i in label2id.items()}

TARGET_COLS = 11


def normalize_box(box, width, height):
    """
    Normalize bbox pixel -> [0,1000]
    """
    x1, y1, x2, y2 = box

    x1 = int(1000 * x1 / width)
    x2 = int(1000 * x2 / width)
    y1 = int(1000 * y1 / height)
    y2 = int(1000 * y2 / height)

    # clamp an toàn
    x1 = max(0, min(1000, x1))
    y1 = max(0, min(1000, y1))
    x2 = max(0, min(1000, x2))
    y2 = max(0, min(1000, y2))

    # tránh bbox rỗng
    if x2 <= x1:
        x2 = min(1000, x1 + 1)
    if y2 <= y1:
        y2 = min(1000, y1 + 1)

    return [x1, y1, x2, y2]


def load_dataset():
    samples = []

    for fname in os.listdir(DATA_FOLDER):
        if not (fname.endswith(".tsv") or fname.endswith(".csv")):
            continue

        image_name = fname.rsplit(".", 1)[0] + ".jpg"
        image_path = os.path.join(IMAGE_FOLDER, image_name)

        if not os.path.exists(image_path):
            print(f"⚠️ Missing image: {image_name}")
            continue

        img = Image.open(image_path)
        width, height = img.size

        words, boxes, labels = [], [], []

        with open(os.path.join(DATA_FOLDER, fname), encoding="utf-8") as f:
            reader = csv.reader(f, delimiter=",")

            for row_idx, row in enumerate(reader, start=1):
                if len(row) != TARGET_COLS:
                    continue

                text = row[-2].strip()
                label = row[-1].strip()

                if label not in label2id:
                    raise ValueError(
                        f"❌ Unknown label '{label}' in {fname} line {row_idx}"
                    )

                try:
                    x_coords = list(map(int, row[1:9:2]))
                    y_coords = list(map(int, row[2:9:2]))
                except ValueError:
                    print(f"⚠️ Invalid coords in {fname} line {row_idx}")
                    continue

                raw_box = [
                    min(x_coords),
                    min(y_coords),
                    max(x_coords),
                    max(y_coords),
                ]

                norm_box = normalize_box(raw_box, width, height)

                words.append(text)
                boxes.append(norm_box)
                labels.append(label2id[label])

        if len(words) == 0:
            print(f"⚠️ Empty text: {fname}")
            continue

        samples.append({
            "image_path": image_path,
            "words": words,
            "boxes": boxes,
            "labels": labels,
        })

    print(f"Loaded {len(samples)} samples")
    return samples
```

```python
samples = load_dataset()

# check bbox
for s in samples[:3]:
    for b in s["boxes"][:5]:
        assert all(0 <= v <= 1000 for v in b), b

# check label
for s in samples[:3]:
    for l in s["labels"][:5]:
        assert 0 <= l < len(label2id), l

print("✅ BBox & labels OK")
```

```python
train_samples, temp_samples = train_test_split(
    samples, test_size=0.2, random_state=42
)

val_samples, test_samples = train_test_split(
    temp_samples, test_size=0.5, random_state=42
)

print(len(train_samples), len(val_samples), len(test_samples))
```

```python
import torch
from torch.utils.data import Dataset
from transformers import LayoutLMv3Processor

processor = LayoutLMv3Processor.from_pretrained(
    "microsoft/layoutlmv3-base",
    apply_ocr=False
)

class InvoiceDataset(Dataset):
    def __init__(self, samples):
        self.samples = samples

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        sample = self.samples[idx]
        image = Image.open(sample["image_path"]).convert("RGB")

        encoding = processor(
            image,
            sample["words"],
            boxes=sample["boxes"],
            word_labels=sample["labels"],
            truncation=True,
            padding="max_length",
            return_tensors="pt",
        )

        return {k: v.squeeze(0) for k, v in encoding.items()}
```

```python
train_dataset = InvoiceDataset(train_samples)
val_dataset   = InvoiceDataset(val_samples)
test_dataset  = InvoiceDataset(test_samples)
```

```python
from transformers import LayoutLMv3ForTokenClassification, Trainer, TrainingArguments

model = LayoutLMv3ForTokenClassification.from_pretrained(
    "microsoft/layoutlmv3-base",
    num_labels=len(label2id),
    id2label=id2label,
    label2id=label2id,
)
```

```python
training_args = TrainingArguments(
    output_dir="./layoutlm_no_bio",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=5,
    weight_decay=0.01,
    logging_steps=50,
    load_best_model_at_end=True,
    report_to="none",
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
)
```

```python
trainer.train()
```

```python
metrics = trainer.evaluate(test_dataset)
print(metrics)
```

```python
from transformers import LayoutLMv3Processor, LayoutLMv3ForTokenClassification
import torch

BASE_MODEL = "microsoft/layoutlmv3-base"
CHECKPOINT = "/content/layoutlm_no_bio/checkpoint-920"

device = "cuda" if torch.cuda.is_available() else "cpu"

processor = LayoutLMv3Processor.from_pretrained(
    "microsoft/layoutlmv3-base",
    apply_ocr=False
)

model = LayoutLMv3ForTokenClassification.from_pretrained(
    CHECKPOINT,
    id2label=id2label,
    label2id=label2id,
).to(device)

model.eval()
```

```python
def normalize_box(box, w, h):
    return [
        int(1000 * box[0] / w),
        int(1000 * box[1] / h),
        int(1000 * box[2] / w),
        int(1000 * box[3] / h),
    ]

def run_inference(image_path, words, boxes):
    image = Image.open(image_path).convert("RGB")

    encoding = processor(
        image,
        words,
        boxes=boxes,
        return_tensors="pt",
        truncation=True,
        padding="max_length"
    )

    word_ids = encoding.word_ids(0)

    encoding = {k: v.to(device) for k, v in encoding.items()}

    with torch.no_grad():
        outputs = model(**encoding)

    pred_ids = outputs.logits.argmax(-1)[0].cpu().tolist()

    results = []
    used = set()

    for idx, word_id in enumerate(word_ids):
        if word_id is None or word_id in used:
            continue

        label = id2label[pred_ids[idx]]
        results.append((words[word_id], label))
        used.add(word_id)

    return results
```

```python
def simple_group_to_text(input_ids, label_ids, tokenizer, id2label):
    fields = defaultdict(list)
    current_label = None
    current_tokens = []

    tokens = tokenizer.convert_ids_to_tokens(input_ids)

    for token, label_id in zip(tokens, label_ids):
        if label_id == -100: continue
        label = id2label[label_id]

        # Bỏ qua nhãn rác
        if label in ["OTHER", "O"]:
            if current_label is not None:
                text = tokenizer.convert_tokens_to_string(current_tokens).strip()
                if text: fields[current_label].append(text)
                current_label = None
                current_tokens = []
            continue

        # Logic gom nhóm
        if label != current_label:
            if current_label is not None:
                text = tokenizer.convert_tokens_to_string(current_tokens).strip()
                if text: fields[current_label].append(text)
            current_label = label
            current_tokens = [token]
        else:
            current_tokens.append(token)

    if current_label is not None and current_tokens:
        text = tokenizer.convert_tokens_to_string(current_tokens).strip()
        if text: fields[current_label].append(text)

    return {k: " ".join(v) for k, v in fields.items()}

# TÍNH KHOẢNG CÁCH (THAY THẾ JIWER)
def get_edit_distance(ref, hyp):
    """Tính số lượng ký tự cần sửa đổi (Levenshtein Distance)"""
    ref = ref.strip()
    hyp = hyp.strip()
    n, m = len(ref), len(hyp)

    if n == 0: return m
    if m == 0: return n

    dp = np.zeros((n + 1, m + 1), dtype=int)
    for i in range(n + 1): dp[i][0] = i
    for j in range(m + 1): dp[0][j] = j

    for i in range(1, n + 1):
        for j in range(1, m + 1):
            cost = 0 if ref[i - 1] == hyp[j - 1] else 1
            dp[i][j] = min(dp[i - 1][j] + 1, dp[i][j - 1] + 1, dp[i - 1][j - 1] + cost)

    return dp[n][m]

token_preds = np.argmax(preds.predictions, axis=-1)
total_chars = defaultdict(int)
total_edits = defaultdict(int)

print("Đang đánh giá lại (Fixed)...")

for i in range(len(test_dataset)):
    # 1. Prediction
    pred_fields = simple_group_to_text(
        input_ids=test_dataset[i]["input_ids"],
        label_ids=token_preds[i],
        tokenizer=tokenizer,
        id2label=id2label
    )

    # 2. Ground Truth
    gt_fields = simple_group_to_text(
        input_ids=test_dataset[i]["input_ids"],
        label_ids=preds.label_ids[i],
        tokenizer=tokenizer,
        id2label=id2label
    )

    # 3. Tính toán
    all_keys = set(gt_fields.keys()) | set(pred_fields.keys())

    for field in all_keys:
        val_gt = gt_fields.get(field, "")
        val_pred = pred_fields.get(field, "")

        dist = get_edit_distance(val_gt, val_pred)

        total_edits[field] += dist
        total_chars[field] += len(val_gt)
```
